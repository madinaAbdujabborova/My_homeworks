```python
import pandas as pd
```

```python
df = pd.read_csv("asthma_disease_data.csv")
df
```

Patient ID: from 5034 to 7425, Bemorlarni ID si
Age: from 5 to 80 years
Gender: 0 - Male, 1 - Female
EthnicityL: 0: Caucasian
            1: African American
            2: Asian
            3: Other
Education level: 0: None
                 1: High School
                 2: Bachelor's
                 3: Higher
BMI: Body mass index. from 15 to 40 (18.5 to 24.9 - healthy, 25.0 to 29.9 - overweight, 30.0 or higher - obese)
Smoking: 1 - Yes, 0 - No
PhysicalActivity: Weekly physical activity in hours, ranging from 0 to 10.
DietQuality: Diet quality score, ranging from 0 to 10.
SleepQuality: Sleep quality score, ranging from 4 to 10.

### Astma kasalligi bilan kasallanganlarni o'rtacha yoshi va Body mass indexi

### Astma bilan kasallanganlarning modasi(Eng kop takrorlanuvchi yoshi)

```python
# Standard og'ish(yosh va BMI bo'yicha bo'yicha)

df[["Age", "BMI"]].std()
```
