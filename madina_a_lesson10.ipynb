```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("index 2.csv")
df
```

```python
df.shape
```

### Problem: Plot the count of each coffee type sold.

```python
sns.countplot(data = df, x="coffee_name", hue="coffee_name")
plt.xticks(rotation = 90)
plt.show()
```

#### Hulosa: Eng ko'p sotiladigan cofe type bu Americano with milk(200), Latte(tahminan178). Eng kam sotiladigani esa Cocoa va Exspresso 

### Problem: Compare the average amount of money spent for each coffee type.

```python
spend_money = df.groupby("coffee_name")["money"].mean()
spend_money
```

```python
sns.barplot(data=df, x="coffee_name", y="money", hue="coffee_name", estimator="mean")

plt.xticks(rotation = 90)
plt.show()
```

#### Boya ko'rgan count plotda Cocoa eng kam natijani ko'rsatgan bo'lsa ham eng ko'p sarflanadigan pul aynan Cocoa ga tegishli. Ammo Espressoga sarflangan pul halli ham eng kam miqdroda.

### Problem: Visualize the distribution of money spent for each coffee type.

```python
sns.set= ( "whitegrid")
sns.boxplot(data=df, y="coffee_name", x="money", hue="coffee_name")
```

```python
df["coffee_name"].value_counts()
```

### Problem: Visualize the distribution of money spent for each coffee type with more detail.

```python
sns.violinplot(data = df, y="coffee_name", x="money", hue = "coffee_name")
```

#### Violent plot bizga ko'proq ma'lumotlani ko'rsata oladi. Misol uchun Hot chocolate typeiga to'langan pul hammasinikidan ham ko'proq ekan Buni o'rtadagi oq chiziqdan ko'rishimiz mumkin. Cappuccino va Latte uchun asosiy sale esa 37.5. 

### Problem: Display all individual transactions of money spent for each coffee type.

```python
sns.stripplot(data=df, y = "coffee_name", x="money", hue="coffee_name")
```

#### Problem: Display all individual transactions of money spent for each coffee type while avoiding overlap.

```python
sns.swarmplot(data=df, y = "coffee_name", x="money", hue="coffee_name")
```

#### ma'lum bir coffe uchun sarflangan pullar typei ko'p bo'lmaganligi sababli stripplot va swarmplotlar chiroyli chiqmadi.
