```python
df['DEBT_MDN_nnumeric'] = pd.to_numeric(df['DEBT_MDN'], errors='coerce')
df[df["DEBT_MDN_nnumeric"].isna()]['DEBT_MDN'].unique()
```

#### Endilikda numerik tipida yangi ustunga tenglab oldim. Va Aynan nimani hisobiga NaN lar oshib ketayotganini tekshirdim. Natijada Ba'zi bir kataklarda son yoki NaN ni o'rniga "PrivacySuppressed" yozuvi bo'lgan va ularning barchasi yangi ustunda NaN

```python
df['DEBT_MDN_nnumeric'].hist()
```

```python
df['DEBT_MDN_nnumeric'].describe()
```

```python
df['DEBT_MDN_nnumeric'] = df['DEBT_MDN_nnumeric'].fillna(df['DEBT_MDN_nnumeric'].median())
df['DEBT_MDN_nnumeric'].isnull().sum()
```

#### numericga aylantirib olgan ustunimni osha ustunning medianasi bilan to'ldirib oldim va nullar qoldimi o'qmi tekshirib ko'rdim.


```python
df['DEBT_MDN_nnumeric'].describe()
```

```python
df['DEBT_MDN_nnumeric'].hist()
```

HIstogrammadan ko'rinib turiptiki bizda mediana ancha oshib ketgan

#### Huddi shu amalni endi boshqa numeric ustunlar uchun ham bajarib ko'ramiz.

```python
df['COMPL_RPY_5YR_RT_numeric'] = pd.to_numeric(df['COMPL_RPY_5YR_RT'], errors='coerce')
df['COMPL_RPY_5YR_RT_numeric'].isnull().sum()
```

#### Avval bizda 994 ta bo'lgan NaN lar endilikda 2251 taga o'zgarib ketti. Agar bu holatda NaN lar o'rniga mean yo median qo'yib chiqsak. Ma'lumotlarimizda katta o'zgarishlar sodir bo'lishi mumkin. Shuning uchun endi boshqa ustunlar bilan bo'g'likligi bormi yo'qmi tekshirib ko'ramiz.

```python
df.groupby(by=['Degree_Type', 'Ownership'])['COMPL_RPY_5YR_RT_numeric'].agg(['count', 'mean', 'median'])
```

```python
df.groupby(by=['REGION', 'Ownership'])['COMPL_RPY_5YR_RT_numeric'].agg(['count', 'mean', 'median'])
```

#### Region va ownership bilan aloqasi bor deb o'yladim, endi Shu ikkala ustunning o'rtachasi bilan NaN larni to'ldirib chiqaman.

```python
df['COMPL_RPY_5YR_RT_numeric'] = df.groupby(by=['REGION', 'Ownership', 'Degree_Type'])['COMPL_RPY_5YR_RT_numeric'].transform(lambda x: x.fillna(x.mean()))
df
```

```python
# fillna dan oldin Histagramma
pd.to_numeric(df['COMPL_RPY_5YR_RT'], errors='coerce').hist()
```

```python
df['COMPL_RPY_5YR_RT_numeric'].hist()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

pd.to_numeric(df['COMPL_RPY_5YR_RT'], errors='coerce').plot(kind='hist', ax=ax, alpha=0.5, color='blue', bins=20, label='Numeric')
df['COMPL_RPY_5YR_RT_numeric'].plot(kind='hist', ax=ax, alpha=0.5, color='green', bins=20, label='Fillna_Mean')

ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.legend()
plt.grid()

plt.title('Histograms of NaN, Fillna_Mean')
plt.show()
```

## 2. Ma'lumotlarni grafiklar orqali tahlil qilish (joinplot, pairplot, vhkz)

```python
df['COMPL_RPY_5YR_RT'] = pd.to_numeric(df['COMPL_RPY_5YR_RT'], errors='coerce')
df['DEBT_MDN'] = pd.to_numeric(df['DEBT_MDN'], errors='coerce')
df['MEDIAN_HH_INC'] = pd.to_numeric(df['MEDIAN_HH_INC'], errors='coerce')

df
```

```python
g = sns.FacetGrid(df, col="HIGHDEG")
g.map(sns.boxplot, 'Tuition')
```

#### HIhgh degreeni barcha darajalari uchun mediana va ma'lumotlarning 50 foizi deayli 2000 atrofida. Eng qimmat tuition esa 1 - va 2 - darajalarda kuzatilgan

```python
g = sns.FacetGrid(df, col="REGION")
g.map(plt.scatter, 'Tuition', 'SAT_AVG_ALL')
```

#### Regionlar hisobida esa Eng ko'p va zichrog'i 5 - region, reggion 0 da esa eng kami

```python
sns.lmplot(data=df, x='Tuition', y='SAT_AVG_ALL', col="HIGHDEG", fit_reg=False)
```

#### Sar_avg_all(nimaligini bilmadim.) ko'rsatgichi darajadan darajaga o'tgan sari ko'payib bormoqda va 4 - darajada eng ko'pi ko'rinyapti

```python
g = sns.PairGrid(df, vars = ['AVGFACSAL','Tuition'])
g = g.map(sns.scatterplot)
```

#### AvgFascal ko'rsatkichi va Tuitionni o'zaro bir biriga mosligini scacttterplot orqali ko'rilyapti. 2000 atrofida zich bo'lganligini ko'rishimiz mumkin

```python
g = sns.PairGrid(df, vars = ['AVGFACSAL','Tuition'])
g = g.map_diag(sns.histplot)
g = g.map_offdiag(sns.scatterplot)
```

HUddi yuqoridagini hist va scatterplot holatidagi ko'rinishi

```python
sns.pairplot(df.query('REGION < 3'), 
             vars=['AVGFACSAL','Tuition', 'SAT_AVG_ALL'],
             hue="REGION", palette="husl", 
             plot_kws={'alpha': 0.5})
```

### 0, 1, 2 - regionlarning AVGFACSAL, Tution va SAT_AVG_ALL ko'rsatkichlarining o'zaro nisbatini pairplot orqali ko'rinishi
