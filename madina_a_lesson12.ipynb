```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('fastfood.csv')
df
```

```python
df.info()
```

```python
df = df.drop(columns=['restaurant', 'item', 'salad'])
df
```

```python
sns.histplot(data = df, x='calories', kde = True)
```

```python
mean = df['calories'].mean()
min = df['calories'].min()
max = df['calories'].max()
std = df['calories'].std()
```

```python
df["new_calories"] =(df['calories']-mean)/std
df
```

```python
sns.histplot(data = df, x='new_calories', kde = True)
```

```python
df['new_calories'].describe()
```

#### Bu standartization bo'lib mean(o'rtacha) 0 ga va std esa 1 ga teng bo'lib qoladi. qiymatlar o'zgargan bo'lsa ham ular orasidagi o'zaro nisbat o'zgargani yo'q grafik ham deyarli o'zgarmagan

```python
# Yuqoridagi ishni butun data frame uchun bajarish
(df-df.mean())/df.std()
```

```python
# min_max_scaling
df['min_max_calories'] = (df['calories'] - min) / (max - min)
df
```

```python
sns.histplot(data = df, x='min_max_calories', kde=True)
```

#### Bu min max scaling formulasi bo'lib qiymatlar 0 dan 1 gacha qiymatlar oralig'ida yotadi. Histplot o'zgargani yo'q


```python
# BUtun dataframe uchun min max scaling qilish
(df - df.min()) / (df.max() - df.min())
```
